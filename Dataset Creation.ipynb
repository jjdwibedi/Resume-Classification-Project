# 1] Import Dependencies

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)

#regular expression
import re

#extract data 
import glob
import docx
import doc

import spacy 
from spacy.matcher import Matcher
# load pre-trained model
nlp = spacy.load('en_core_web_sm')
import warnings
warnings.filterwarnings("ignore")
```

# 2] Creating dataset from all the resumes 

```python
path1 = 'F:\\Data Science\\Projects\\NLP\\Resumes\\SQL Developer Lightning insight'
path2 = 'F:\\Data Science\\Projects\\NLP\\Resumes\\React Developer'
path3 = 'F:\\Data Science\\Projects\\NLP\\Resumes\\Peoplesoft resumes'
path4 = 'F:\\Data Science\\Projects\\NLP\\Resumes\\workday resumes'

sql=glob.glob(path1 + "/*.docx")
react=glob.glob(path2 + "/*.docx")
ppl=glob.glob(path3 + "/*.docx")
work=glob.glob(path4 + "/*.docx")
```

```python
def readtxt(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
Resume1=[]
Category1=[]
for filename in sql:
    dummy1=readtxt(filename)
    Resume1.append(dummy1)
    Category1.append("SQL Developer")
    SQL = (pd.DataFrame(Resume1,Category1).reset_index())
    SQL.columns = ['Category','Resume']


Resume2=[]
Category2=[]
for filename in react:
    dummy2=readtxt(filename)
    Resume2.append(dummy2)
    Category2.append("React JS Developer")
    reactjs = (pd.DataFrame(Resume2,Category2).reset_index())
    reactjs.columns = ['Category','Resume']
    

Resume3=[]
Category3=[]
for filename in ppl:
    dummy3=readtxt(filename)
    Resume3.append(dummy3)
    Category3.append("Peoplesoft")
    Peoplesoft = (pd.DataFrame(Resume3,Category3).reset_index())
    Peoplesoft.columns = ['Category','Resume']


Resume4=[]
Category4=[]
for filename in work:
    dummy4=readtxt(filename)
    Resume4.append(dummy4)
    Category4.append('Work Day')
    workday = (pd.DataFrame(Resume4,Category4).reset_index())
    workday.columns = ['Category','Resume']


df = pd.concat([SQL,reactjs, Peoplesoft, workday],axis=0).reset_index()
df = df.drop(columns='index',axis=0)
df['Resume']= df['Resume']
df.head()
```

# 3] Creating functions to extract data from Resume column

```python
def extract_name(text):
    nlp_text = nlp(text)

    matcher = Matcher(nlp.vocab)
    pattern = [{'POS':'PROPN','POS':'PROPN'}]
    matcher.add('PERSON', [pattern])
    matches = matcher(nlp_text)
    for match_id, start, end in matches:
        span = nlp_text[start:end]
        return span.text

df['Name'] = df['Resume'].apply(extract_name)
df['Name'] = df['Name'].str.strip()
df
```

```python
def extract_skills(text, category):
    skills = []
    if category == 'SQL Developer':
        predefined_skills = ['SQL', 'ETL', 'Stored Procedures', 'Database Management', 'Data Analysis', 
                             'Query Optimization', 'Performance Tuning', 'Data Modeling', 
                             'Indexing', 'Data Warehousing']
    
    elif category == 'Work Day':
        predefined_skills = ['Workday', 'HCM', 'Integration', 'Consultant', 'Business Processes', 
                             'Configuration', 'Report Writing', 'Security Administration', 'Absence Management', 'Benefits Administration']
    
    elif category == 'Peoplesoft':
        predefined_skills = ['Peoplesoft', 'HRMS', 'Payroll Management', 'PeopleTools',
                             'Application Designer', 'PeopleCode', 'Component Interface', 'SQR', 
                             'Integration Broker', 'Workflow']
    
    elif category == 'React JS Developer':
        predefined_skills = ['React JS', 'UI Development', 'MERN stack','JavaScript', 'JSON',
                              'React','HTML', 'CSS', 'RESTful APIs', 'React Router', 'Redux', 
                             'Webpack', 'Testing (Jest/Enzyme)']
    else:
        predefined_skills = []

    for skill in predefined_skills:
        pattern = r'\b{}\b'.format(re.escape(skill), flags=re.IGNORECASE)
        if re.search(pattern, text, flags=re.IGNORECASE):
            skills.append(skill)

    return ', '.join(skills[:10])

# Apply the function to the DataFrame based on the Category column
df['Skills'] = df.apply(lambda x: extract_skills(x['Resume'], x['Category']), axis=1)
df
```

```python
import nltk
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))

# Education Degrees
EDUCATION = ['BE','B.E.', 'B.E', 'BS','B.S','B.Com','BCA','ME','M.E', 
             'M.E.', 'M.S','B.com','10','10+2','BTECH', 'B.TECH', 'M.TECH', 
             'MTECH', 'SSC', 'HSC', 'C.B.S.E','CBSE','ICSE', 'X', 'XII','10th',
             '12th',' 10th',' 12th','Bachelor of Arts in Mathematics',
             'Master of Science in Analytics','Bachelor of Business Administration',
             'Major: Business Management']

def extract_education(text):
    nlp_text = nlp(text)
    nlp_text = [sent.text.strip() for sent in nlp_text.sents]
    edu = {}
    for index, t in enumerate(nlp_text):
        for tex in t.split():
            tex = re.sub(r'[?|$|.|!|,]', r'', tex)
            if tex in EDUCATION and tex not in STOPWORDS:
                edu[tex] = t + nlp_text[index + 1]

    education = []
    for key in edu.keys():
        year = re.search(re.compile(r'(((20|19)(\d{2})))'), edu[key])
        if year:
            education.append((key, ''.join(year[0])))
        else:
            education.append(key)
    return education

df['Education'] = df['Resume'].apply(extract_education)
df
```

```python
def extract_date_of_birth(resume_text):
    # Define patterns for common date formats
    date_patterns = [
        r"\b(\d{1,2})[/.-](\d{1,2})[/.-](\d{4})\b",  # dd/mm/yyyy or dd-mm-yyyy or dd.mm.yyyy
        r"\b(\d{4})[/.-](\d{1,2})[/.-](\d{1,2})\b",  # yyyy/mm/dd or yyyy-mm-dd or yyyy.mm.dd
    ]

    # Iterate over the date patterns and search for a match in the resume text
    for pattern in date_patterns:
        match = re.search(pattern, resume_text)
        if match:
            # Return the matched date
            return match.group()

    return None

# Apply the extract_date_of_birth function to the "Resume" column
df['DOB'] = df['Resume'].apply(extract_date_of_birth)
df
```

```python
def extract_experience(text):
    experience_pattern = r'\b(?!2012|100)(?:(?:0?\.\d+|\d+\.\d+)|[1-9]\d?(?:\.\d+)?)(?:\s*(?:years?|yrs?)\s*)?(?:\s*\d+(?:\.\d+)?\s*(?:months?|mos?)\s*)?(?:\s*(?:of\s)?experience)?'
    experience = re.search(experience_pattern, text, re.IGNORECASE)
    if experience:
        experience_value = re.findall(r'\d+(?:\.\d+)?', experience.group())
        # Exclude numbers greater than 26
        if float(experience_value[0]) <= 26:
            return experience_value[0]
    return ''

df['Experience_Years'] = df['Resume'].apply(extract_experience)
df
```

```python
def extract_previous_work_experience(resume_text):
    # Define a pattern to extract previous work experience
    experience_pattern = r'(\d+(\.\d+)?\s*(?:years?|yrs?)\s*(?:of\s+experience)?)[\s-]*(.*)'

    # Search for the experience pattern in the resume text
    match = re.search(experience_pattern, resume_text, re.IGNORECASE)

    if match:
        return match.group(3).strip()  # Extract the third group (previous work experience)
    else:
        return 'Not Found'

# Apply the extract_previous_work_experience function to the "Resume" column
df['Work_Experience'] = df['Resume'].apply(extract_previous_work_experience)
df
```

```python
def extract_phone_number(text):
    phone_pattern = r'(\+91)?\d{10}'
    phone = re.search(phone_pattern, text)
    if phone:
        return phone.group()
    else:
        return '-'

df['Phone'] = df['Resume'].apply(extract_phone_number)
df
```

```python
def extract_email(text):
    email_pattern = r'\b[A-Za-z0-9._%+-]+@(?:[A-Za-z0-9-]+\.)+[A-Za-z]{2,}\b'
    email = re.search(email_pattern, text)
    if email:
        return email.group()
    else:
        return '-'

df['Email_ID'] = df['Resume'].apply(extract_email)
df
```

```python
def extract_links(resume_text):
    link_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    links = re.findall(link_pattern, resume_text)
    if links:
        return links
    else:
        return '-'

# Apply the extract_links function to the "Resume" column
df['Links'] = df['Resume'].apply(extract_links)
df
```

```python
def clean_resume_data(resume_text):
    resume_text = re.sub('http\S+\s*', ' ', resume_text)  # remove URLs
    resume_text = re.sub('#\S+', '', resume_text)  # remove hashtags
    resume_text = re.sub('@\S+', ' ', resume_text)  # remove mentions
    resume_text = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resume_text)  # remove punctuations
    resume_text = re.sub(r'[^\x00-\x7f]', ' ', resume_text) 
    resume_text = re.sub('\s+', ' ', resume_text)  # remove extra whitespace
    return resume_text

df['Resume'] = df['Resume'].apply(clean_resume_data)
df
```
